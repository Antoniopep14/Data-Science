### planteamiento del problema
sugerir juegos a nuevos jugadores de nintendo

solo juegos que puedan interesarle

solo 1 juego por mes


```python
import pandas as pd 
data_juegos = pd.read_csv("juegos-ml.csv")
data_juegos
```

```python
#la ventaja de este set de datos es que no tenemos duplicados, nulos,
#incorrectos ni irrelevantes por lo que no tenemos que limpiarlos
#ahora tenemos que pasarle los datos de entrada(que queremos analizar)
#y salida(los que esperamos)
X = data_juegos.drop(columns=("juegos"))
#el metodo drop quita la columna designada y crea una tabla nueva
X
```

```python
#ahora tenemos que obtener el set de datos que queremos tener
#para ello vamos a acceder a una columna en particular
y = data_juegos["juegos"]
y
```

```python
#vamos a intentar hacer una prediccion con 14 a√±os
from sklearn.tree import DecisionTreeClassifier
modelo = DecisionTreeClassifier()
modelo.fit(X.values, y) #el primero es el modelo base y el segundo es lo esperado
#pero los datos deben de ir sin los nombre de las columnas
predicciones = modelo.predict([[14, 0], [29, 0], [30, 0]])
predicciones
```

```python
#ahora anterirormente le pasamos todos los datos
#pero lo correcto es divirlos en datos de entrenamiento datos
from sklearn.model_selection import train_test_split
##para comparar modelos importamos lo siguiente
from sklearn.metrics import accuracy_score

data_juegos = pd.read_csv("juegos-ml.csv")
X = data_juegos.drop(columns=("juegos"))
y = data_juegos["juegos"]
#agreggamos la variable de entrenamiento
X_entrenar, X_prueba, y_entrenar, y_prueba = train_test_split(X.values, y, test_size=0.2)#aqui le pasamos lo datos base, los de resultado
#y un argumento nombrado con el % de los datos que queremos usar como prueba
#y recibe un numero de 0 a 1 (0.2 es 20%)
#esto nos devulve una tupla, y lo que haremos es desempaquetar esa tupla
#y asignar esos valores a una variable

modelo = DecisionTreeClassifier()
modelo.fit(X_entrenar, y_entrenar)
predicciones = modelo.predict(X_prueba)
predicciones
#ahora lo que tenemos que hacer es comparar los resultados de los modelos
#de prueba con el modelo de prediccion real
puntaje = accuracy_score(y_prueba, predicciones)
puntaje#el valor de abajo va a ir cambiando porque tenemos muy pocos datos
#para aumentar la exactitud podriamos aumentar el % de pruebas
```

```python
#ahora tocaria persistir el modelo, para no tener que probarlo cada
#vez que queramos usarlo y podemas seguir usandolo posteriormente
#para ello tenemos que quitar todo lo de las pruebas y agregar
import joblib
#quitar las predicciones y agregar
joblib.dump(modelo, 'recomendador-juegos.joblib' )#esto va a recibir el modelo, nombre con el que lo
# guardaremos y guardarlo en la ruta que le indiquemos
#despues de guardarlo ya podemos solo cargarlo con
modelo = joblib.load('recomendador-juegos.joblib')
predicciones = modelo.predict([[14, 0]])
```
